```python
# numeric features for clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
numeric_features = ['Car_price', 'Battery_cap', 'Drive_range', 'Power', 'Charge_time', 'Boot_space', 'Top_speed']
X = df[numeric_features]

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determining the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the elbow method to find the optimal number of clusters
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

```python
# Based on the elbow method, let's assume 3 clusters
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
df['Segment'] = kmeans.fit_predict(X_scaled)
```

```python
# Visualizing Segments 
feature1, feature2 = 'Car_price', 'Drive_range'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x=feature1, y=feature2, hue='Segment', palette='viridis')
plt.title('Segmentation Analysis')
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.show()
```

```python
# Profiling and Describing Segments
segment_stats = df.groupby('Segment')[numeric_features].describe()
print(segment_stats)
```

```python
additional_df = pd.read_csv('EVIndia.csv')
#numeric features for clustering in the additional dataset
numeric_features = ['PriceRange']  # Adjust feature names
X = additional_df[numeric_features]
```

```python
# Converting the 'PriceRange' feature to a numeric value (remove 'Cr' and convert to float)
X['PriceRange'] = X['PriceRange'].str.replace(' Cr', '').astype(float)
```

```python
# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determining the optimal number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the elbow method to find the optimal number of clusters
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal Clusters (Additional Dataset)')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

```python
# Based on the elbow method, let's assume 3 clusters for the additional dataset
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
additional_df['Segment'] = kmeans.fit_predict(X_scaled)
```

```python
# Profiling and Describing Segments
segment_stats = additional_df.groupby('Segment')['PriceRange'].describe()
print(segment_stats)
```

```python
# Visualizing Segments with Different Colors
plt.figure(figsize=(10, 6))
sns.scatterplot(data=additional_df, x='PriceRange', y='Segment', hue='Segment', palette='viridis', legend='full')
plt.title('Segmentation Analysis (Additional Dataset)')
plt.xlabel('Price Range (in Crores)')
plt.ylabel('Segment')
plt.show()
```

```python
#the distribution of vehicle types in each segment:
plt.figure(figsize=(8, 6))
sns.countplot(data=additional_df, x='VehicleType', hue='Segment', palette='viridis')
plt.title('Distribution of Vehicle Types in Segments')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.legend(title='Segment')
plt.show()
```

```python
import re

# Function to extract numerical values from 'PriceRange' strings
def extract_price(price_str):
    match = re.search(r'\d+\.\d+', price_str)
    if match:
        return float(match.group())
    else:
        return 0.0  # Return 0 if no numerical value found

additional_df['Price'] = additional_df['PriceRange'].apply(extract_price)

# Grouping 'Price' into broader categories
def categorize_price(price):
    if price < 0.2:
        return 'Low'
    elif price < 0.5:
        return 'Medium'
    else:
        return 'High'

additional_df['PriceCategory'] = additional_df['Price'].apply(categorize_price)

# Analyzing the distribution of price categories for electric vehicles using a countplot
plt.figure(figsize=(10, 6))
sns.countplot(data=additional_df, y='PriceCategory', palette='viridis')
plt.title('Distribution of Price Categories for Electric Vehicles')
plt.xlabel('Count')
plt.ylabel('Price Category')
plt.show()

```
